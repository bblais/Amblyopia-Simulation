```python
sfnames=['sims/2025-02-24//md 20 neurons dog 0.asdf']
```

```python
strong_i=1
weak_i=0

R=Results(sfnames[0])
```

```python
kk=R.max_SF
t=R.t

=kk.mean(axis=1)  # average across neurons, at the end of a seq, for each channel
S=kk.std(axis=1)
N=kk.shape[1]
K=1+20/N**2
=K*S/np.sqrt(N)


errorbar(t,[:,strong_i],yerr=[:,strong_i],fmt='b.',lw=1,ecolor='k')
errorbar(t,[:,weak_i],yerr=[:,weak_i],fmt='m.',lw=1,ecolor='k')
ylabel('Max Spatial Frequency')
```
