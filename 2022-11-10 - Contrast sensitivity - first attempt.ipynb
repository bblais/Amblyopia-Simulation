```python
def (x): ## unit gaussian
    return exp(-pi*x**2)

def (u): ## fourier transform is the same form
    return exp(-pi*u**2)

def a(x,a=1): ## scaled unit gaussian
    return 1/abs(a)*(x/a)

def a(u,a=1): ## fourier transform is the same form
    return (a*u)

# 2D

def 2a(x,y,a=1): ## scaled unit gaussian
    return a(x,a)*a(y,a)

def 2a(ux,uy,a=1): ## fourier transform is the same form
    return a(ux,a)*a(uy,a)

```

```python
def myfft(x,y):
    import numpy as np
    yf=np.fft.fft(y)
    
    dx=x[1]-x[0]
    N=len(yf)
    
    k=np.fft.fftfreq(N, d=dx)
    
    yf=np.fft.fftshift(yf)  # unscramble the fft
    k=np.fft.fftshift(k)    # unscramble the frequencies

    return k,yf
```

```python
k,yf=myfft(x,y)
```

```python
plot(k,abs(yf))
xlim([-1,1])
```
