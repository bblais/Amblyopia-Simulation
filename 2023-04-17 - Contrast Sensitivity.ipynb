```python
exponent=linspace(-4,2,500)
x=10**exponent
```

```python
def u(x,kc,rc,ks,rs):
    return kc*rc*sqrt(pi)*exp(-(x/rc)**2)-ks*rs*sqrt(pi)*exp(-(x/rs)**2)
```

```python
plot(x,u(x,
         kc=6,
         rc=1,
         ks=.1,
         rs=18))
xscale('log')
yscale('log')
```
