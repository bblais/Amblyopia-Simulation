![[blob_convolution_example_fsig_20.png]]{#fig:dichopic_blob}
> The dichoptic masks are produced by taking random circular blobs (lower left), convolving them a Gaussian filter of a specified size (middle), resulting in the circular blobs blending into the background smoothly at the edges on the scale of the filter.  
---

![[mask_filter_examples_fsigs.png]]{#fig:dichopic_filter_size}
> The dichoptic masks for several different filter sizes.  The larger the filter, the larger the overlap in the patterns presented to the two eyes.  For a very sharp mask (upper left) patterns are nearly all to either the left or the right eye.  For a wide mask (lower right) most patterns are presented to both eyes.
---


![[mask_filter_example_fsig_20.png]]{#fig:dichopic_filter_image}
> An example of a dichoptic mask $\sigma_f=20$ applied to one of the images.  The mask is shown in the top-center, and the two masked images are in the lower left and right.  One can see areas where there is some overlap as well as areas where the pattern is only in one of the eyes due to the application of the mask.

![[recovery contrast.png]]{#fig:dODI_constrast}
> The rate of recovery, defined as the change in the ODI vs time,  for the contrast treatment.  $\text{contrast = 1}$ is normal vision, whereas  $\text{contrast = 0}$ is the strong-eye receives no input.  The rate of recovery decreases as the balance between the two eyes is made more equal (as $\text{contrast}$ tends toward $1$).
----

![[recovery contrast and mask 1.png]]{#fig:dODI_constrast_mask}
> The rate of recovery, defined as the change in the ODI vs time,  for the contrast treatment with dichoptic mask.  $\text{contrast = 1}$ is normal vision, whereas  $\text{contrast = 0}$ is the strong-eye receives no input.  The mask with $f=10$ has the least overlap between the eyes, whereas the mask with $f=90$ has the most overlap between the eyes.  The mask with the largest overlap between the eyes, the effect of contrast approaches that of no-mask.  As the overlap between the eyes is decreased, by making the mask filter smaller, the rate of recovery is increased for small contrast but *decreased* for larger contrast.  There is even an indication that a small-filter mask combined with nearly equal input to both eyes *will make the deficit worse*.
